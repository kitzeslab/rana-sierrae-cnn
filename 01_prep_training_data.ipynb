# create table of one-hot labels from Raven annotations

```python
import pandas as pd
from opensoundscape.annotations import BoxedAnnotations
from glob import glob
from pathlib import Path
```

```python
data_dir='/media/emu/datasets/annotated/rana_sierrae_2022/'
file_df=pd.read_csv('resources/audio_and_raven_files.csv')
```

```python
valid_labels = ['A','B','C','D','E','X']
```

```python
all_annotations = []
for af, rf in zip(file_df.audio,file_df.raven):
    ba = BoxedAnnotations.from_raven_file(rf,annotation_column='Annotation',audio_file=af) #allow us to pass a list of raven files (and audio files?)
    ba.df = ba.df.dropna() 
    if len(ba.df)>0:
        clip_df = ba.one_hot_clip_labels(classes=['A','B','C','V','X'],
                               full_duration=10,
                               clip_duration=2,
                               clip_overlap=1.5,
                               min_label_overlap=0.2,
                               min_label_fraction=0.5,
                            ).astype(int)
        clip_df['file']=af
        clip_df = clip_df.reset_index().set_index(['file','start_time','end_time'])
        all_annotations.append(clip_df)
labels = pd.concat(all_annotations)
```
